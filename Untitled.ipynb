```python
import numpy as np

def nw(x, y, match = 1, mismatch = 1, gap = 1):
    nx = len(x)
    ny = len(y)
    # Optimal score at each possible pair of characters.
    F = np.zeros((nx + 1, ny + 1))
    F[:,0] = np.linspace(0, -nx * gap, nx + 1)
    F[0,:] = np.linspace(0, -ny * gap, ny + 1)
    # Pointers to trace through an optimal aligment.
    P = np.zeros((nx + 1, ny + 1))
    P[:,0] = 3
    P[0,:] = 4
    # Temporary scores.
    t = np.zeros(3)
    for i in range(nx):
        for j in range(ny):
            #if x[i] == y[j]:
            t[0] = F[i,j] + y[j]- 0.3
            #else:
                #t[0] = F[i,j] - mismatch
            #t[1] = F[i,j+1] - gap
            t[2] = F[i+1,j] - gap
            tmax = np.max(t)
            F[i+1,j+1] = tmax
            if t[0] == tmax:
                P[i+1,j+1] += 2
            if t[1] == tmax:
                P[i+1,j+1] += 3
            if t[2] == tmax:
                P[i+1,j+1] += 4
    # Trace through an optimal alignment.
    i = nx
    j = ny
    rx = []
    ry = []
    while i > 0 or j > 0:
        if P[i,j] in [2, 5, 6, 9]:
            rx.append(x[i-1])
            ry.append(y[j-1])
            i -= 1
            j -= 1
        elif P[i,j] in [3, 5, 7, 9]:
            rx.append(x[i-1])
            ry.append('-')
            i -= 1
        elif P[i,j] in [4, 6, 7, 9]:
            rx.append('-')
            ry.append(y[j-1])
            j -= 1
    # Reverse the strings.
    #rx = ''.join(rx)[::-1]
    #ry = ''.join(ry)[::-1]
    return [rx, ry]
```

```python
def gotoh(seq1, seq2,gap_open=0, gap_extend=-0.3):
    len1, len2 = len(seq1), len(seq2)
    M = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    Ix = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    Iy = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    print('hello')
    for i in range(1, len1 + 1):
        M[i][0] = gap_open + (i - 1) * gap_extend
        Ix[i][0] = gap_open + (i - 1) * gap_extend
        Iy[i][0] = float('-inf')

    for j in range(1, len2 + 1):
        M[0][j] = gap_open + (j - 1) * gap_extend
        Iy[0][j] = gap_open + (j - 1) * gap_extend
        Ix[0][j] = float('-inf')

    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            score = (seq1[i-1] ) * seq2[j-1]
           
            Ix[i][j] = max(M[i - 1][j] + gap_open + gap_extend, Ix[i - 1][j] + gap_extend)
            Iy[i][j] = max(M[i][j - 1] + gap_open + gap_extend, Iy[i][j - 1] + gap_extend)
            M[i][j] = max(M[i - 1][j - 1] + score, Ix[i][j], Iy[i][j])

    print(np.array(M))
    align1, align2 = [], []
    i, j = len1, len2
    while i > 0 and j > 0:
        print('hello')
        score = (seq1[i-1] - 0.3) * seq2[j-1]
        if M[i][j] == M[i - 1][j - 1] + score:
            align1.append(seq1[i - 1])
            align2.append(seq2[j - 1])
            i -= 1
            j -= 1
            print('match')
        elif M[i][j] == Ix[i][j]:
            align1.append(seq1[i - 1])
            align2.append('-')
            i -= 1
            print('mismatch')
        elif M[i][j] == Iy[i][j]:
            align1.append('-')
            align2.append(seq2[j - 1])
            j -= 1
            print('mismatch')

    while i > 0:
        align1.append(seq1[i - 1])
        align2.append('-')
        i -= 1

    while j > 0:
        align1.append('-')
        align2.append(seq2[j - 1])
        j -= 1

    return [align1, align2]

```

```python
def gotoh_plus(seq1, seq2,gap_open=0, gap_extend=-0.3):
    len1, len2 = len(seq1), len(seq2)
    M = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    Ix = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    Iy = [[0] * (len2 + 1) for _ in range(len1 + 1)]
    
    for i in range(1, len1 + 1):
        M[i][0] = gap_open + (i - 1) * gap_extend
        Ix[i][0] = gap_open + (i - 1) * gap_extend
        Iy[i][0] = float('-inf')
        
        
    for j in range(1, len2 + 1):
        M[0][j] = gap_open + (j - 1) * gap_extend
        Iy[0][j] = gap_open + (j - 1) * gap_extend
        Ix[0][j] = float('-inf')

    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            score = (seq1[i-1]) * seq2[j-1]

            Ix[i][j] = min(M[i - 1][j] + gap_open + gap_extend, Ix[i - 1][j] + gap_extend)
            Iy[i][j] = min(M[i][j - 1] + gap_open + gap_extend, Iy[i][j - 1] + gap_extend)
            
            M[i][j] = min(M[i - 1][j - 1] + score, Ix[i][j], Iy[i][j])

    align1, align2 = [], []
    i, j = len1, len2
    while i > 0 and j > 0:
        score = (seq1[i-1]) * seq2[j-1]
        if M[i][j] == M[i - 1][j - 1] + score:
            align1.append(seq1[i - 1])
            align2.append(seq2[j - 1])
            i -= 1
            j -= 1
        elif M[i][j] == Ix[i][j]:
            align1.append(seq1[i - 1])
            align2.append('-')
            i -= 1
        elif M[i][j] == Iy[i][j]:
            align1.append('-')
            align2.append(seq2[j - 1])
            j -= 1

    while i > 0:
        align1.append(seq1[i - 1])
        align2.append('-')
        i -= 1

    while j > 0:
        align1.append('-')
        align2.append(seq2[j - 1])
        j -= 1

    return [align1, align2]

```
